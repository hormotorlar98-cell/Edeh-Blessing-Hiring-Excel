```python
# Step 1: Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Step 2: Load Excel file from GitHub
url = "https://github.com/hormotorlar98-cell/Edeh-Blessing-Hiring-Excel"  # Replace with your file URL
df = pd.read_excel(url, sheet_name='Sheet1')

# Step 3: Inspect the data
print("Original Data:")
print(df)

# Step 4: Clean the data
exp_map = {
    "one":1, "two":2, "three":3, "four":4, "five":5,
    "six":6, "seven":7, "eight":8, "nine":9, "ten":10,
    "eleven":11, "twelve":12
}

df['experience'] = df['experience'].map(exp_map)

# Fill missing numeric values with column mean
for col in ['experience', 'test_score', 'interview_score', 'salary']:
    df[col] = df[col].fillna(df[col].mean())

# Step 5: Features and target
X = df[['experience', 'test_score', 'interview_score']]
y = df['salary']

# Step 6: Train Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Step 7: Predict salaries for new candidates
new_candidates = pd.DataFrame({
    'experience': [2, 12],
    'test_score': [9, 10],
    'interview_score': [6, 10]
})

predicted_salaries = model.predict(new_candidates)

Predicted Salaries for New Candidates:
Candidate 1: $48000.00
Candidate 2: $90000.00


print("\nPredicted Salaries for New Candidates:")
for i, salary in enumerate(predicted_salaries, start=1):
    print(f"Candidate {i}: ${salary:.2f}")
```
