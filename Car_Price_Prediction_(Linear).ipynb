```python
# GROUP 5 Car Price Prediction
# Name	IDS
# 1. Endurance Iloba	WTF/2025/3546
# 2. Mercy Mwangi	WTF/2025/5169
# 3. Betselot Berhanu	WTF/2025/8987
# 4. Edeh Blessing	WTF/2025/6076
# 5. Ola Ifeoluwa	WTF/2025/3108
# 6. Zintle Mgwelo	WTF/2025/6802
# 7. Susan Omondi	WTF/2025/8557
#8. Winfred Mutitu	WTF/2025/5164
#9. Vivian Chikaodili	WTF/2025/6730


import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_excel('/content/drive/MyDrive/carprices 3.xlsx')
df.head()
```

```python
# Converting Text to Numbers
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
x = df.drop('Sell Price($)', axis=1)
y = df['Sell Price($)']

#Created a Linear Regression model
reg = LinearRegression()

#Trained the model using the dataset
reg.fit(x,y)
```

```python
# Making predictions
def predict_price(mileage, age, car_model):
    row = { "Mileage": mileage, "Age(yrs)": age,
        "Car Model_BMW X5": 1 if car_model == "BMW X5" else 0,
        "Car Model_Mercedez Benz C class": 1 if car_model == "Mercedez Benz" else 0 }
    return reg.predict(pd.DataFrame([row]))[0]
```

```python
#1. Predict price of a mercedez benz that is 4 yr old with mileage 4500

predict_price(45000, 4,"Mercedez Benz")
```

```python
#2. Predict price of a BMW X5 that is 7 yr old with mileage 86000

predict_price(86000, 7, "bmw X5")
```

```python
# 3. Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

reg.score(x, y)
```
